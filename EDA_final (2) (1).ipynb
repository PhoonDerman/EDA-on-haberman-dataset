# DATA EXPLORATION



```python
import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt

import numpy as np

data = pd.read_csv('Downloads\haberman.csv')
print(data.head)
```

```python

print(data.shape) 
print(data.columns)

print(data['status'].value_counts())
```

##### Observation

There are 306 rows and 4 columns with column names as "age", "year", "nodes" and "status".

Non of the columns have missing value and are integers.

Total value under status 1 - 225

Total value under status 2- 81

The data is imbalanced.

#### Observation 

Above code gives us information about the data like counts, mean, std etc.

### 2D Scatter Plot 



```python

sns.set_style("whitegrid");

sns.FacetGrid(data, hue= "status", height= 4).map(plt.scatter, "age", "year").add_legend();
plt.title("scatter plot (age vs year)")
plt.show();
```

```python

sns.set_style("whitegrid");

sns.FacetGrid(data, hue = "status", height = 4).map(plt.scatter, "year", "nodes").add_legend();
plt.title("scatter plot (year vs node)")
plt.show();
```

```python

sns.set_style("whitegrid");

sns.FacetGrid(data, hue = "status", height = 4).map(plt.scatter, "age", "nodes").add_legend();
plt.title("scatter plot (age vs node)")
plt.show();
```

##### Observation

1. Majority of the patients have less than 5 nodes and also, patients with age less than 40 have much higher chance of surviving and have less nodes detected.

2. Patients above the age of 50 and 10 + nodes have less chance of suriving. The majority of patients were between age 40 and 65.



## Pair-plots

```python

plt.close();

sns.set_style("whitegrid");

sns.pairplot(data, hue = "status", height = 4, vars =['age','year','nodes']);

plt.show()
```

##### OBSERVATION


1. Most patients have 0 nodes or close to zero and patients with less number of nodes have higher chance survival and people with higher nodes have smaller chance of survival. 

2. Survival rate is higher before 1965 and less after year 1965. 




### Univariate Analysis


```python

live_longer = data.loc[data["status"] == 1]

live_shorter = data.loc[data["status"] == 2]

plt.plot(live_longer["age"], np.zeros_like(live_longer['age']), 'o')

plt.plot(live_shorter["age"], np.zeros_like(live_shorter['age']), 'o')

plt.show()
```


##### Observation

Data not clear since there are too many points too close to eachother and survival rate of 35age is better.



**Histogram**

```python

sns.FacetGrid(data, hue = "status", height =4).map(sns.distplot, "nodes").add_legend();
plt.title("Histogram Of nodes")
#when using displot the graph seperated into 3 different graph, so using distplot instead


plt.show();
```

##### Observation

1. Age 35< have higher survival rate, age 40-50 have highest survival rate and age 65+ have lower.

2. Age 40-70 have majority of patients.

3. Year and age have higher overlaping class lebels.



**CDF and PDF** 


```python
counts, bin_edges = np.histogram(live_longer["nodes"], bins = 10, density = True)

pdf = counts / (sum(counts))
cdf = np.cumsum(pdf)
print(bin_edges)
print("probability density function")
print(pdf)
print("Cumulative distribution function")
print(cdf)



plt.plot(bin_edges[1:],pdf, label ="pdf")
plt.plot(bin_edges[1:], cdf, label = "cdf")
plt.xlabel('nodes')
plt.title("Pdf and Cdf of status 1")
plt.legend()

```

### Observations

* About 90+% patients who survived have 0 to 10 nodes
* Patients who not survived contains more number of nodes

```python
counts, bin_edges = np.histogram(live_shorter["nodes"], bins= 10, density = True)

pdf = counts /(sum(counts))
cdf = np.cumsum(pdf)

print(pdf)
print(bin_edges)
print("Probability density function")
print(pdf)
print("Cumulative distribution function")
print(cdf)

plt.plot(bin_edges[1:], pdf, label="pdf")
plt.plot(bin_edges[1:], cdf, label= "cdf")

plt.xlabel("nodes")
plt.title("Pdf and Cdf of status 2")
plt.legend()
```

### Observation

* About 70+% patients who not survived have 0 to 10 nodes
* Patients who not survived contains more number of nodes


### Mean, Variance and Std-dev


```python
print("Means : ")

print (np.mean(live_longer["nodes"]))

print (np.mean(live_shorter["nodes"]))

print("\nVariance : ")

print (np.var(live_longer["nodes"]))

print (np.var(live_shorter["nodes"]))

print("\nStandard Deviation : ")

print (np.std(live_longer["nodes"]))
print (np.std(live_shorter["nodes"]))
```


##### Observation

1. Patients who survived had an average of 2.7 nodes.

2. Patients who didn't survive had an average of 7.4 nodes which is significantly higher than survived patients.



### Median, Percentile, Quantile, IQR, MAD


```python
print("Medians:")

print(np.median(live_longer['nodes']))

print(np.median(live_shorter['nodes']))



print("\nQuantiles:")

print(np.percentile(live_longer['nodes'], np.arange(0, 100, 25)))

print(np.percentile(live_shorter['nodes'], np.arange(0, 100, 25)))



print("\n50th percentile")

print(np.percentile(live_longer['nodes'], 50))

print(np.percentile(live_shorter['nodes'], 50))

```


##### Observation

1. Survived patients had 0 nodes on 75% and 3 nodes on 100%.

2. Less survived had 1 node on 25%, 4 nodes on 75% and 11 nodes on 100%.
